```python
import pandas as pd
```

```python
vine_file = '/Users/michaelmanthey/Documents/Bootcamp/Module_16_BigData/Challenge/vine_table.csv'

vine_df = pd.read_csv(vine_file)
vine_df
```

```python
# new df to hold total votes greater than 20
filtered_df = vine_df.loc[vine_df['total_votes'] >= 20]
#filtered_df = vine_df.loc['total_votes'] >= 20
filtered_df
```

```python
helpful_df = filtered_df.loc[vine_df['helpful_votes'] / vine_df['total_votes'] >= .5]
helpful_df
```

```python
vine_updated_df = helpful_df.loc[helpful_df['vine'] == 'Y']
vine_updated_df

non_vine_updated_df = helpful_df.loc[helpful_df['vine'] == 'N']
```

```python
# total votes
# paid
total_votes_paid = vine_updated_df['total_votes'].sum()
#unpaid
total_votes_unpaid = non_vine_updated_df['total_votes'].sum()

#five_stars
#paid
five_star_filter_paid = vine_updated_df['star_rating'] >= 5
five_star_votes_paid = vine_updated_df['star_rating'].count()

#unpaid
five_star_filter_unpaid = non_vine_updated_df['star_rating'] >= 5
five_star_votes_unpaid = non_vine_updated_df['star_rating'].count()

# percentage of five stars
five_star_percent_paid = five_star_votes_paid / total_votes_paid
five_star_percent_unpaid = five_star_votes_unpaid / total_votes_unpaid
```

```python
total_votes_unpaid
```

```python
# Summary DFs
paid_summary = pd.DataFrame({
    "Total Votes - Paid": total_votes_paid,
    "Five Star Votes - Paid": five_star_votes_paid,
    "Five Star Percent - Paid": five_star_percent_paid
}, index=[1])

unpaid_summary = pd.DataFrame({
    "Total Votes - Unpaid": total_votes_unpaid,
    "Five Star Votes - Unpaid": five_star_votes_unpaid,
    "Five Star Percent - Unpaid": five_star_percent_unpaid
}, index=[1])
```
